# 온도인자 15개 Correlationshipe 분석후 feature selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn           as sb
import scipy.stats       as sp
```

```python
path = './factor table/total_temp_factor_table.csv'

df = pd.read_csv(path)
```

```python
df
```

```python
feature = df.columns
drop_ = [6, 10,11,13] # input_rms, pre_max, pre_min , pre_rms
feature_ = feature.drop(df.columns[drop_])
feature = feature_[4:]
feature.shape

# feature = df.columns[4:]
```

```python
train_co = df[feature]

train_co.corr()
pc =  train_co.corr()
plt.figure(figsize = (15,15))

sb.heatmap(pc, cmap="YlGnBu",annot = True)
plt.show()
```

# 절대값 0.9 넘는애들만 살리고 나머지 0

```python
threshold = 0.3
feature_num = feature.shape[0]
feature_temp = pc.copy() # 피쳐 이름 살리기 위해 원본 결과 복사
for i in range(feature_num):
    for j in range(feature_num):
        temp_pc = pc.iloc[i,j]
        if (abs(temp_pc) < threshold):
            feature_temp.iloc[i,j] = 0
    
feature_temp.shape # (특징개수,특징개수)
```

```python
plt.figure(figsize = (15,15))
sb.heatmap(feature_temp,annot = True)
plt.show()
```
