## 라이브러리 가져오기

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib
```

```python
Data = pd.read_csv('./GripTestData.csv', sep = ',') 

Data_X = Data.iloc[:, :8]
Data_Y = Data.iloc[:, 8:]
```

## XX YY XY 인자 상관계수 계산

```python
## X-Y 인자

# 상관계수 값 배열
PearsonArray = np.zeros((Data_X.shape[1]-1, Data_Y.shape[1]))

# 각각의 상관계수 계산
for i in range(1,Data_X.shape[1]): ## 원단 No 제외
    for j in range(Data_Y.shape[1]):
        TempX = Data_X.iloc[:,i]
        TempY = Data_Y.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonArray[i-1, j] = TempTest.corr(method='pearson').iloc[0,1]
        
PearsonArray = pd.DataFrame(PearsonArray)

Row_Ind = Data_X.columns[1:]
Column_Ind = Data_Y.columns
PearsonArray.index = Row_Ind
PearsonArray.columns = Column_Ind

PearsonArray = PearsonArray.T

PearsonArray.to_csv('./PearsonCoeffResult/PearsonCoeffDataXY.csv', sep = ',', header=None, index=None)
PearsonArray
```

```python
## X-X 인자

# 상관계수 값 배열
PearsonArray = np.zeros((Data_X.shape[1]-1, Data_X.shape[1]-1))

# 각각의 상관계수 계산
for i in range(1,Data_X.shape[1]): ## 원단 No 제외
    for j in range(1,Data_X.shape[1]):
        TempX = Data_X.iloc[:,i]
        TempY = Data_X.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonArray[i-1, j-1] = TempTest.corr(method='pearson').iloc[0,1]
        
        if i >= j:
            PearsonArray[i-1, j-1] = 0  # 대칭되는 값 삭제
        
PearsonArray = pd.DataFrame(PearsonArray)

Row_Ind = Data_X.columns[1:]
Column_Ind = Data_X.columns[1:]
PearsonArray.index = Row_Ind
PearsonArray.columns = Column_Ind

PearsonArray.to_csv('./PearsonCoeffResult/PearsonCoeffDataXX.csv', sep = ',', header=None, index=None)
PearsonArray
```

```python
## Y-Y 인자

# 상관계수 값 배열
PearsonArray = np.zeros((Data_Y.shape[1], Data_Y.shape[1]))

# 각각의 상관계수 계산
for i in range(Data_Y.shape[1]): ## 원단 No 제외
    for j in range(Data_Y.shape[1]):
        TempX = Data_Y.iloc[:,i]
        TempY = Data_Y.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonArray[i, j] = TempTest.corr(method='pearson').iloc[0,1]
        
        if i >= j:
            PearsonArray[i, j] = 0  # 대칭되는 값 삭제
        
PearsonArray = pd.DataFrame(PearsonArray)

Row_Ind = Data_Y.columns
Column_Ind = Data_Y.columns
PearsonArray.index = Row_Ind
PearsonArray.columns = Column_Ind

PearsonArray.to_csv('./PearsonCoeffResult/PearsonCoeffDataYY.csv', sep = ',', header=None, index=None)
PearsonArray
```

## Y 인자 중 처짐량 관련 제거 - YY

```python
# Y 인자 중 열 제거 (5, 6, 8, 9, 11, 12)
Data_Y_Deleted = Data_Y.copy()
# del Data_Y_Deleted['center deflection max(mm)','center deflection min(mm)', 'max deg L', 'min deg L', 'max deg R', 'min deg R']
Data_Y_Deleted.drop(['center deflection max(mm)','center deflection min(mm)', 'max deg L', 'min deg L', 'max deg R', 'min deg R'], axis='columns', inplace=True)

## Y-Y 인자 
# 상관계수 값 배열
PearsonArray = np.zeros((Data_Y_Deleted.shape[1], Data_Y_Deleted.shape[1]))

# 각각의 상관계수 계산
for i in range(Data_Y_Deleted.shape[1]): ## 원단 No 제외
    for j in range(Data_Y_Deleted.shape[1]):
        TempX = Data_Y_Deleted.iloc[:,i]
        TempY = Data_Y_Deleted.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonArray[i, j] = TempTest.corr(method='pearson').iloc[0,1]
        
        if i >= j:
            PearsonArray[i, j] = 0  # 대칭되는 값 삭제
        
PearsonArray = pd.DataFrame(PearsonArray)

Row_Ind = Data_Y_Deleted.columns
Column_Ind = Data_Y_Deleted.columns
PearsonArray.index = Row_Ind
PearsonArray.columns = Column_Ind

PearsonArray.to_csv('./PearsonCoeffResult/PearsonCoeffDataYY_Deleted.csv', sep = ',', header=None, index=None)
PearsonArray
```

## Y 인자 중 처짐량 관련 제거 - XY

```python
## X-Y 인자

# 상관계수 값 배열
PearsonArray = np.zeros((Data_X.shape[1]-1, Data_Y_Deleted.shape[1]))

# 각각의 상관계수 계산
for i in range(1,Data_X.shape[1]): ## 원단 No 제외
    for j in range(Data_Y_Deleted.shape[1]):
        TempX = Data_X.iloc[:,i]
        TempY = Data_Y_Deleted.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonArray[i-1, j] = TempTest.corr(method='pearson').iloc[0,1]
        
PearsonArray = pd.DataFrame(PearsonArray)

Row_Ind = Data_X.columns[1:]
Column_Ind = Data_Y_Deleted.columns
PearsonArray.index = Row_Ind
PearsonArray.columns = Column_Ind

PearsonArray = PearsonArray.T

PearsonArray.to_csv('./PearsonCoeffResult/PearsonCoeffDataXY_Deleted.csv', sep = ',', header=None, index=None)
PearsonArray
```

## 정렬된 값들 저장

```python
X_Ind = Data_X.columns[1:]
Y_Ind = Data_Y.columns
Y_Ind_Del = Data_Y_Deleted.columns

PearsonArrayXX = pd.read_csv('./PearsonCoeffResult/PearsonCoeffDataXX.csv', header=None)
PearsonArrayXX.index = X_Ind
PearsonArrayXX.columns = X_Ind

PearsonArrayXY = pd.read_csv('./PearsonCoeffResult/PearsonCoeffDataXY.csv', header=None)
PearsonArrayXY.index = Y_Ind
PearsonArrayXY.columns = X_Ind

PearsonArrayXY_Deleted = pd.read_csv('./PearsonCoeffResult/PearsonCoeffDataXY_Deleted.csv', header=None)
PearsonArrayXY_Deleted.index = Y_Ind_Del
PearsonArrayXY_Deleted.columns = X_Ind

PearsonArrayYY = pd.read_csv('./PearsonCoeffResult/PearsonCoeffDataYY.csv', header=None)
PearsonArrayYY.index = Y_Ind
PearsonArrayYY.columns = Y_Ind

PearsonArrayYY_Deleted = pd.read_csv('./PearsonCoeffResult/PearsonCoeffDataYY_Deleted.csv', header=None)
PearsonArrayYY_Deleted.index = Y_Ind_Del
PearsonArrayYY_Deleted.columns = Y_Ind_Del
```

## XY 인자 상관관계 정렬, (삭제된 Y 인자도)

```python
# 상관계수 내림차순 정렬
PearsonForSort = np.zeros((PearsonArrayXY.shape[0]*PearsonArrayXY.shape[1], 4))
PearsonForSort = pd.DataFrame(PearsonForSort)
for i in range(PearsonArrayXY.shape[1]):
    for j in range(PearsonArrayXY.shape[0]):
        TempX = Data_X.iloc[:,i+1]
        TempY = Data_Y.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonForSort.iloc[PearsonArrayXY.shape[0]*i+j, 0] = TempX.name
        PearsonForSort.iloc[PearsonArrayXY.shape[0]*i+j, 1] = TempY.name
        PearsonForSort.iloc[PearsonArrayXY.shape[0]*i+j, 2] = TempTest.corr(method='pearson').iloc[0,1]
        PearsonForSort.iloc[PearsonArrayXY.shape[0]*i+j, 3] = np.abs(TempTest.corr(method='pearson').iloc[0,1])
        
PearsonForSort.columns = ['X factor', 'Y factor', 'PearsonCoeff', 'PearsonCoeff(abs)']
PearsonForSort = PearsonForSort.sort_values(by = 'PearsonCoeff(abs)', ascending = False)
XYPearsonSorted = PearsonForSort.iloc[:,:3]
XYPearsonSorted.to_csv('./PearsonCoeffResult/XYPearsonSorted.csv', sep = ',', index=None)
XYPearsonSorted
```

```python
# 상관계수 내림차순 정렬
PearsonForSort = np.zeros((PearsonArrayXY_Deleted.shape[0]*PearsonArrayXY_Deleted.shape[1], 4))
PearsonForSort = pd.DataFrame(PearsonForSort)
for i in range(PearsonArrayXY_Deleted.shape[1]):
    for j in range(PearsonArrayXY_Deleted.shape[0]):
        TempX = Data_X.iloc[:,i+1]
        TempY = Data_Y_Deleted.iloc[:,j]
        TempTest = pd.concat([TempX,TempY], axis = 1)
        
        PearsonForSort.iloc[PearsonArrayXY_Deleted.shape[0]*i+j, 0] = TempX.name
        PearsonForSort.iloc[PearsonArrayXY_Deleted.shape[0]*i+j, 1] = TempY.name
        PearsonForSort.iloc[PearsonArrayXY_Deleted.shape[0]*i+j, 2] = TempTest.corr(method='pearson').iloc[0,1]
        PearsonForSort.iloc[PearsonArrayXY_Deleted.shape[0]*i+j, 3] = np.abs(TempTest.corr(method='pearson').iloc[0,1])
        
PearsonForSort.columns = ['X factor', 'Y factor', 'PearsonCoeff', 'PearsonCoeff(abs)']
PearsonForSort = PearsonForSort.sort_values(by = 'PearsonCoeff(abs)', ascending = False)
XYPearsonSorted = PearsonForSort.iloc[:,:3]
XYPearsonSorted.to_csv('./PearsonCoeffResult/XYPearsonSorted_Deleted.csv', sep = ',', index=None)
XYPearsonSorted
```

## XX YY XY 인자 상위 4개 플로팅

```python
# 실험데이터 가져오기
Data = pd.read_csv('./GripTestData.csv', sep = ',') 

# 정렬된 데이터 csv 가져오기
XXPearsonSorted = pd.read_csv('./PearsonCoeffResult/XXPearsonSorted.csv', sep = ',')
YYPearsonSorted = pd.read_csv('./PearsonCoeffResult/YYPearsonSorted.csv', sep = ',')
XYPearsonSorted = pd.read_csv('./PearsonCoeffResult/XYPearsonSorted.csv', sep = ',')
```

```python
# 상위 4개
plt.figure(1)
for i in range(4):
    # 각각의 인자 Name 추출
    Xname = XXPearsonSorted.iloc[i,0]
    Yname = XXPearsonSorted.iloc[i,1]
    pearson = str(round(XXPearsonSorted.iloc[i,2], 3))
    
    x = Data.iloc[:,Data.columns.get_loc(Xname)]
    y = Data.iloc[:,Data.columns.get_loc(Yname)]
    xname = x.name
    yname = y.name
    x = np.array(x)
    y = np.array(y)
    line_fitter = LinearRegression()
    line_fitter.fit(x.reshape(-1, 1), y)
    
    plt.subplot(2,2,i+1)
    plt.plot(x,line_fitter.predict(x.reshape(-1,1)), label='Linear Regression')
    plt.scatter(x, y, label='Actual Data', s=25, c='r')
    plt.title("Rank"+str(i+1)+" (Pearson's R:"+pearson+')')

    plt.xlabel(xname)
    plt.ylabel(yname)
    plt.grid()
    plt.show()
    
    
# 하위 4개
plt.figure(2)
for i in range(4):
    # 각각의 인자 Name 추출
    Xname = XXPearsonSorted.iloc[-i-1,0]
    Yname = XXPearsonSorted.iloc[-i-1,1]
    pearson = str(round(XXPearsonSorted.iloc[-i-1,2], 3))
    
    x = Data.iloc[:,Data.columns.get_loc(Xname)]
    y = Data.iloc[:,Data.columns.get_loc(Yname)]
    xname = x.name
    yname = y.name
    x = np.array(x)
    y = np.array(y)
    line_fitter = LinearRegression()
    line_fitter.fit(x.reshape(-1, 1), y)
    
    plt.subplot(2,2,i+1)
    plt.plot(x,line_fitter.predict(x.reshape(-1,1)), label='Linear Regression')
    plt.scatter(x, y, label='Actual Data', s=25, c='r')
    plt.title("Rank"+str(i+1)+" (Pearson's R:"+pearson+')')

    plt.xlabel(xname)
    plt.ylabel(yname)
    plt.grid()
    plt.show()
```
