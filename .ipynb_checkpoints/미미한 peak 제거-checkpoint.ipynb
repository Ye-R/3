## peak 찾기

```python
# 앞뒤로 쭉 읽다가 증가하는 포인트와 peak 사이의 값 비교(100개)
peak_dp     = []
nonepeak_dp = []
for m in range(100):
    print("-"*60)
    print("dp%d"%(m+1))
    exec("temp_thickness = thickness_dp%d"%(m+1))
    temp_peak_index = []

    temp_thickness_reverse = temp_thickness[::-1]
    
    for i in range(temp_thickness.shape[0]-2):
        if temp_thickness.iloc[i,1]<temp_thickness.iloc[i+1,1]:
            if temp_thickness.iloc[i+1,1]>temp_thickness.iloc[i+2,1]:
                temp_peak_index = np.append(temp_peak_index, i+1)

    try:
        for j in range (temp_peak_index.shape[0]):
            print("peak %d index : "%(j+1),int(temp_peak_index[j]))
            #피크 기준 오른쪽
            for k in range(temp_thickness.shape[0]-int(temp_peak_index[j])-1):
                if temp_thickness.iloc[int(temp_peak_index[j])-1+k,1] > temp_thickness.iloc[int(temp_peak_index[j])+k,1]:
                    if temp_thickness.iloc[int(temp_peak_index[j])+k,1] < temp_thickness.iloc[int(temp_peak_index[j])+k+1,1]:
                        print("peak %d right side_rising index : "%(j+1),temp_peak_index[j]+k)
                        right_diff = temp_thickness.iloc[int(temp_peak_index[j]),1]-temp_thickness.iloc[int(temp_peak_index[j])+k,1]
                        print("peak %d right side difference : "%(j+1),right_diff)
        #                 print('\t')
                        break
                if np.min(temp_thickness.iloc[int(temp_peak_index[j]):,1]) == temp_thickness.iloc[-1,1]:
                    print("peak %d right side_not rising_end index : "%(j+1),temp_thickness.shape[0]-1)
                    right_diff = temp_thickness.iloc[int(temp_peak_index[j]),1]-temp_thickness.iloc[temp_thickness.shape[0]-1,1]
                    print("peak %d right side difference : "%(j+1),right_diff)
        #             print('\t')
                    break

            #피크 기준 왼쪽
            a = temp_thickness.shape[0]-int(temp_peak_index[j])
            for k in range(int(temp_peak_index[j])):
                if temp_thickness_reverse.iloc[a+k-2,1] > temp_thickness_reverse.iloc[a+k-1,1]:
                    if temp_thickness_reverse.iloc[a+k-1,1] < temp_thickness_reverse.iloc[a+k,1]:
                        print("peak %d left side_rising index : "%(j+1),temp_peak_index[j]-k)
                        left_diff = temp_thickness.iloc[int(temp_peak_index[j]),1]-temp_thickness.iloc[int(temp_peak_index[j])-k,1]
                        print("peak %d left side difference : "%(j+1),left_diff)
                        print('\t')
                        break
                if np.min(temp_thickness_reverse.iloc[a:,1]) == temp_thickness_reverse.iloc[-1,1]:
                    print("peak %d left side_not rising_end index : "%(j+1),0)
                    left_diff = temp_thickness.iloc[int(temp_peak_index[j]),1]-temp_thickness.iloc[0,1]
                    print("peak %d left side difference : "%(j+1),left_diff)
                    print('\t')
                    break
                    
            ## 여기가 threshold 설정하는데!!!!####
            if right_diff < 0.0001 or left_diff < 0.0001:
                print("No! dp%d peak %d is not peak!"%(m+1,j+1))
                nonepeak_dp = np.append(nonepeak_dp,m+1)
                print("\t")
            else:
                print("Yes! dp%d peak %d is peak!"%(m+1,j+1))
                peak_dp = np.append(peak_dp,m+1)
                print("\t")
    except AttributeError:
        print("No Peak!")
        nonepeak_dp = np.append(nonepeak_dp,m+1)
        print("\t")
        pass
```

```python
peak_dp     = list(set(peak_dp))
nonepeak_dp = list(set(nonepeak_dp))

for i in range(np.shape(peak_dp)[0]):
    if peak_dp[i] in nonepeak_dp:
        nonepeak_dp.remove(peak_dp[i])
```
