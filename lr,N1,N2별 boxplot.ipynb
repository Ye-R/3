```python
Lr005_dataset = np.zeros((60,7))
for i in range(len(Lr005_index)):
    Lr005_dataset[i,:]     = Tr_result_epoch2000.iloc[int(Lr005_index[i]),:]
    Lr005_dataset[i+15,:]  = Tr_result_epoch5000.iloc[int(Lr005_index[i]),:]
    Lr005_dataset[i+30,:]  = Tr_result_epoch10000.iloc[int(Lr005_index[i]),:]
    Lr005_dataset[i+45,:]  = Tr_result_epoch20000.iloc[int(Lr005_index[i]),:]

Lr005_dataset = pd.DataFrame(np.concatenate((header,Lr005_dataset)))
Lr005_dataset = Lr005_dataset.rename(columns=Lr005_dataset.iloc[0])
Lr005_dataset = Lr005_dataset.iloc[1:,:]
# pd.DataFrame(Lr005_dataset)
```

```python
Lr01_dataset = np.zeros((60,7))
for i in range(len(Lr01_index)):
    Lr01_dataset[i,:]     = Tr_result_epoch2000.iloc[int(Lr01_index[i]),:]
    Lr01_dataset[i+15,:]  = Tr_result_epoch5000.iloc[int(Lr01_index[i]),:]
    Lr01_dataset[i+30,:]  = Tr_result_epoch10000.iloc[int(Lr01_index[i]),:]
    Lr01_dataset[i+45,:]  = Tr_result_epoch20000.iloc[int(Lr01_index[i]),:]

Lr01_dataset = pd.DataFrame(np.concatenate((header,Lr01_dataset)))
Lr01_dataset = Lr01_dataset.rename(columns=Lr01_dataset.iloc[0])
Lr01_dataset = Lr01_dataset.iloc[1:,:]
# pd.DataFrame(Lr01_dataset)
```
